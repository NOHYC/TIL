# 동영상 저장 - cv2.VideoWriter

- 일련의 프레임을 파일로 저장할 수 있다.
- 일련의 프레임은 모두 크기와 데이터 타입이 같아야한다. => 아니면 저장 안됨

## Fourcc 동영상 파일의 코덱, 압축방식,색상,픽셀포멧 등을 정의하는 정수 값



|cv2.VideoWriter_fourcc(*'DIVX') | DIVX MPEG-4 코덱 |   
|cv2.VideoWriter_fourcc(*'XVID') | XVID MPEG-4 코덱 |   
|cv2.VideoWriter_fourcc(*'FMP4') | FFMPEG MPEG-4 코덱 |   
|cv2.VideoWriter_fourcc(*'X264') | H.264/ AVC 코덱 |   
|cv2.VideoWriter_fourcc(*'MJPG') | motion-JPEG 코덱 |   
 
 
==> 정확한 건 https://www.fourcc.org/codecs.php#letter_m 있다

## cv2.VideoWriter(filenames, fourcc, fps, frameSize, isColor)

-  filename : 비디오 파일 이름 (e.g. 'video.mp4')
-  fourcc : fourcc (e.g. cv2.VideoWriter_fourcc(*'DIVX'))
-  fps : 초당 프레임 수 (e.g. 30)
- frameSize : 프레임 크기. (width, height) 튜플.
- isColor : 컬러 영상이면 True, 그렇지않으면 False. 기본값은 True


```python
cap = cv2.VideoCapture(0)

# 노트북 이미지를 사이즈를 출력하는 코드인데 저장하는 사이즈보다 작을 경우 동영상 저장 x
# cap.set(cv2.CAP_PROP_FRAME_HEIGHT,600)
# cap.set(cv2.CAP_PROP_FRAME_WIDTH,700)
# HEIGHT, WIDTH 가 100,200 일 경우 저장 x 
# 기본적인 것이지만 다양한 딥러닝 모델을 적용하다보면 모델 출력 사이즈가 저장 사이즈 보다 작아서 저장 안되는 경우가 생김

# 동영상 저장 사이즈
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test.avi',fourcc,fps,(width,height))
while cap.isOpened():
    re,frame = cap.read()
    cv2.imshow('notebook',frame)
    out.write(frame)
    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
    

```

# 이미지 대칭
- cv2.flip(이미지 프레임,숫자(상하좌우))
숫자 : 0( 상하 )// 1 ( 좌우 )

```python
cap = cv2.VideoCapture(0)
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test.avi',fourcc,fps,(width,height))
while cap.isOpened():
    re,frame = cap.read()
    frame = cv2.flip(frame,1)
    cv2.imshow('notebook',frame)
    out.write(frame)
    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

# 이미지 확대 축소  (  Image Pyramids )
    
1. cv2.pyrUp(이미지 프레임, dstsize = (width,height) ) 
- 사이즈 업 방식 : Laplacian pyramid: Used to reconstruct an upsampled image from an image lower in the pyramid (with less resolution)
- 0으로 채워진 짝수 행과 열을 새롭게 삽입하고 나서 가우시안 필터를 적용해 주변 픽셀과 비슷하게 만드는 방법으로 크기는 4배 확대
    
   
2. cv2.pyrDown(이미지 프레임, dstsize = (width,height) )
- 사이즈 다운 방식 : Gaussian pyramid: Used to downsample images
- 가우시안 필터를 적용한 뒤 모든 짝수 행과 열을 삭제해 입력 영상의 1/4 크기로 축소
  
    
   
3. **가우시안 분포란?**
> 가우시안 분포(gaussian distribution)란 정규 분포(normal distribution)이라고도 하는데, 평균 근처에 몰려 있는 값들의 개수가 많고 평균에서 멀어질수록 그 개수가 적어지는 분포를 말합니다.( 가우시안 블러링 커널은 중앙값이 가장 크고 중앙에서 멀어질 수록 그 값이 작아짐-> 노이즈 제거효과 굳 )

```python
cap = cv2.VideoCapture(0)
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test.avi',fourcc,fps,(width,height))
while cap.isOpened():
    re,frame = cap.read()
    frame = cv2.pyrDown(frame,(200,100))
    cv2.imshow('notebook',frame)
    frame = cv2.pyrUp(frame,(width,height))
    out.write(frame) # 저장하면 뿌옇게 출력된다~
    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

# 이미지 회전
1. cv2.getRotationMatrix2D(center, angle, scale)
- Parameters:	
- center – 이미지의 중심 좌표
- angle – 회전 각도
- scale – scale factor
2. Affine Transformation
- Affine Transformation은 선의 평행성은 유지가 되면서 이미지를 변환하는 작업입니다. 이동, 확대, Scale, 반전까지 포함된 변환입니다. Affine 변환을 위해서는 3개의 Match가 되는 점이 있으면 변환행렬을 구할 수 있습니다.

```python
# 그냥 만들면 재미없으니 회전시켜봤습니다

cap = cv2.VideoCapture(0)
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test.avi',fourcc,fps,(width,height))

ang = 0

while cap.isOpened():
    re,frame = cap.read()
    
    matrix = cv2.getRotationMatrix2D((width/2,height/2),ang,1)
    frame = cv2.warpAffine(frame, matrix, (width, height))
    
    cv2.imshow('notebook',frame)
    out.write(frame) 
    # 프레임이 진행될 때마다 10도씩 돌아갑니다.
    ang += 10 
    if ang == 180:
        ang = 0
    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

# 이미지 배율
1. cv2.resize(이미지 프레임, dsize =(width, height), interpolation=cv2.INTER_AREA )
- 이미지 픽셀값 변경 영역 보간법으로 이미지 변경 
- 보간법은 이미지의 크기를 변경하는 경우, 변형된 이미지의 픽셀은 추정해서 값을 할당해야 함 ==> 원하는 사이즈로 변경되지 않을 수 있음
2. cv2.resize(src, dsize=(0, 0), fx=0.3, fy=0.7, interpolation=cv2.INTER_LINEAR)
- 결과 이미지 크기가 (0, 0)으로 크기를 설정하지 않은 경우, fx와 fy를 이용하여 이미지의 비율을 조절
- 이미지 비율 x는 0.3배 y는 0.7배 비율로 바꾼다. 


```python
import cv2

src = cv2.imread("Image/dog3.jpg", cv2.IMREAD_COLOR)

dst = cv2.resize(src, dsize=(640, 480), interpolation=cv2.INTER_AREA) # 이미지 픽셀값 변경 영역 보간법으로 이미지 변경
dst2 = cv2.resize(src, dsize=(0, 0), fx=0.3, fy=0.7, interpolation=cv2.INTER_LINEAR)
# 이미지 비율 x는 0.3배 y는 0.7배 비율로 바꾼다. 

cv2.imshow("src", src)
cv2.imshow("dst", dst)
cv2.imshow("dst2", dst2)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# 이미지 이진화
1. cv2.cvtColor(이미지 프레임 , cv2.COLOR_BGR2GRAY) => 컬러를 흑백으로 변환

2. cv2.threshold(src, thresh, maxval, type) → retval, dst ( 사용된 입계값, 이진값 )
- Parameters:	
- src – input image로 single-channel 이미지.(grayscale 이미지)
- thresh – 임계값
- maxval – 임계값을 넘었을 때 적용할 value
- type – thresholding type
> thresholding type은 아래와 같습니다.
>- cv2.THRESH_BINARY
>- cv2.THRESH_BINARY_INV
>- cv2.THRESH_TRUNC
>- cv2.THRESH_TOZERO
>- cv2.THRESH_TOZERO_INV

```python
cap = cv2.VideoCapture(0)
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test.avi',fourcc,fps,(width,height),isColor = False)
## 흑백으로 저장할 때는 isColor 파라미터를 False 안그러면 저장 안됨!

while cap.isOpened():
    re,frame = cap.read()
    
    frame = cv2.cvtColor(frame , cv2.COLOR_BGR2GRAY) # 이것만 적용하면 흑 ~ 백 중간값이 있어서 사물이 잘 보임
    ret,frame = cv2.threshold(frame, 100 , 255, cv2.THRESH_BINARY) # 임계값을 적용해서 완전 흰색 검정색만 보임 지워지는 것들이 있음
    # ret 사용된 입계값
    cv2.imshow('notebook',frame)
    out.write(frame) 

    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

# 이미지 블러
1. cv2.blur(src, ksize) → dst
- Box형태의 kernel을 이미지에 적용한 후 평균값을 box의 중심점에 적용하는 형태
>Parameters:	 
src – Chennel수는 상관없으나, depth(Data Type)은 CV_8U, CV_16U, CV_16S, CV_32F or CV_64F.       
ksize – kernel 사이즈(ex; (3,3))      

2. cv2.GaussianBlur(img, ksize, sigmaX)
-  Gaussian Filter는 Gaussian함수를 이용한 Kernel을 적용합니다.이미지의 Gaussian Noise (전체적으로 밀도가 동일한 노이즈, 백색노이즈)를 제거하는 데 가장 효과적입니다
>Parameters:	
img – Chennel수는 상관없으나, depth(Data Type)은 CV_8U, CV_16U, CV_16S, CV_32F or CV_64F.       
ksize – (width, height) 형태의 kernel size. width와 height는 서로 다를 수 있지만, 양수의 홀수로 지정해야 함.        
sigmaX – Gaussian kernel standard deviation in X direction.        
3. cv2.medianBlur(src, ksize)
- kernel window와 pixel의 값들을 정렬한 후에 중간값을 선택하여 적용합니다. salt-and-pepper noise 제거에 가장 효과적 
>Parameters:	
src – 1,3,4 channel image. depth가 CV_8U, CV_16U, or CV_32F 이면 ksize는 3또는5, CV_8U이면 더 큰 ksize가능      
ksize – 1보다 큰 홀수 

```python
cap = cv2.VideoCapture(0)
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test.avi',fourcc,fps,(width,height))


while cap.isOpened():
    re,frame = cap.read()
    
    frame = cv2.blur(frame,(2,2)) # 커널의 크기가 커질 수록 흐려짐
    cv2.imshow('notebook',frame)
    out.write(frame) 

    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

# 이미지 가장자리 검출
1. cv2.Canny(image, threshold1, threshold2, edges=None, apertureSize=None, L2gradient=None)
- 최대 스레숄드 보다 높은 부분은 확실한 엣지이고 최소 스레숄드보다 낮은 부분은 에지가 아니라고 판단. 사이값은 픽셀의 연결구조를 보고 에지인지 아닌지 판단
- 파라미터
> image는 입력 이미지   
threshold1, threshold2는 최소 스레숄드와 최대 스레숄드   
edges에 Canny 결과를 저장할 변수를 적어줍니다. 파이썬에선 Canny 함수 리턴으로 받을 수 있기 때문에 필요없는 항목   
apertureSize는 이미지 그레디언트를 구할때 사용하는 소벨 커널 크기입니다. 디폴트는 3.     
L2gradient가 True이면 그레디언트 크기를 계산할 때 sqrt{(dI/dx)^2 + (dI/dy)^2}를 사용.    
False라면 근사값인 |dI/dx|+|dI/dy|를 사용합니다.  디폴트값은 False.    

2. cv2.Sobel(그레이스케일 이미지, 정밀도, x방향 미분, y방향 미분, 커널, 배율, 델타, 픽셀 외삽법)
- 파라미터
>정밀도는 결과 이미지의 이미지 정밀도를 의미합니다. 정밀도에 따라 결과물이 달라질 수 있습니다.  
x 방향 미분은 이미지에서 x 방향으로 미분할 값을 설정합니다.   
y 방향 미분은 이미지에서 y 방향으로 미분할 값을 설정합니다.   
커널은 소벨 커널의 크기를 설정합니다. 1, 3, 5, 7의 값을 사용합니다.   
배율은 계산된 미분 값에 대한 배율값입니다.   
델타는 계산전 미분 값에 대한 추가값입니다.   
픽셀 외삽법은 이미지를 가장자리 처리할 경우, 영역 밖의 픽셀은 추정해서 값을 할당해야합니다.   
이미지 밖의 픽셀을 외삽하는데 사용되는 테두리 모드입니다. 외삽 방식을 설정합니다.   
Tip : x방향 미분 값과 y방향의 미분 값의 합이 1 이상이여야 하며 각각의 값은 0보다 커야합니다.    

3. cv2.Laplacian(그레이스케일 이미지, 정밀도, 커널, 배율, 델타, 픽셀 외삽법)를 이용하여 가장자리 검출을 적용합니다.
- 파라미터
>정밀도는 결과 이미지의 이미지 정밀도를 의미합니다. 정밀도에 따라 결과물이 달라질 수 있습니다.    
커널은 2차 미분 필터의 크기를 설정합니다. 1, 3, 5, 7의 값을 사용합니다.   
배율은 계산된 미분 값에 대한 배율값입니다.   
델타는 계산전 미분 값에 대한 추가값입니다.   
픽셀 외삽법은 이미지를 가장자리 처리할 경우, 영역 밖의 픽셀은 추정해서 값을 할당해야합니다.  
이미지 밖의 픽셀을 외삽하는데 사용되는 테두리 모드입니다. 외삽 방식을 설정합니다.  
Tip : 커널의 값이 1일 경우, 3x3 Aperture Size를 사용합니다. (중심값 = -4)  

```python
cap = cv2.VideoCapture(0)
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test.avi',fourcc,fps,(width,height),isColor = False)


while cap.isOpened():
    re,frame = cap.read()
    
    frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    #frame = cv2.Canny(frame,80,255) # 커널의 크기가 커질 수록 흐려짐
    
#     frame_x = cv2.Sobel(frame, cv2.CV_8U, 1, 0, 3) # CV 정밀도 x축 기준 
#     frame_x = cv2.convertScaleAbs(frame_x)
#     frame_y = cv2.Sobel(frame, cv2.CV_8U, 0, 1, 3) # CV 정밀도 y축 기준 
#     frame_y = cv2.convertScaleAbs(frame_y)
#     frame = cv2.addWeighted(frame_x,1,frame_y,1,0)
    
    frame = cv2.Laplacian(frame, cv2.CV_8U, ksize=5)
    
    cv2.imshow('notebook',frame)
    out.write(frame) 

    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

# 복습한걸 써보자!


```python
# 그냥 만들면 재미없으니 회전시켜봤습니다

cap = cv2.VideoCapture(0)
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test.avi',fourcc,fps,(width,height))

ang = 0

while cap.isOpened():
    re,frame = cap.read()
    
    matrix = cv2.getRotationMatrix2D((width/2,height/2),ang,1)
    frame = cv2.warpAffine(frame, matrix, (width, height))
    frame = cv2.pyrDown(frame,(width//2, height//2))
    
    if ang%3 == 1:
        frame = cv2.bitwise_not(frame)
    
    if ang%2 == 1:
        frame = cv2.blur(frame,(20,20))
    
    
    cv2.imshow('notebook',frame)
    frame = cv2.pyrUp(frame,(width,height))
    out.write(frame) 
    ang += 5 
    if ang == 360:
        ang = 0
    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test2.avi',fourcc,fps,(width,height),isColor = False)

num =0
while cap.isOpened():
    re,frame = cap.read()
    
    frame = cv2.cvtColor(frame,cv2.COLOR_BGR2GRAY)
    if num <3:
        frame = cv2.Canny(frame,100,200) 
    elif num< 6:
        frame_x = cv2.Sobel(frame, cv2.CV_8U, 1, 0, 3) # CV 정밀도 x축 기준 
        frame_x = cv2.convertScaleAbs(frame_x)
        frame_y = cv2.Sobel(frame, cv2.CV_8U, 0, 1, 3) # CV 정밀도 y축 기준 
        frame_y = cv2.convertScaleAbs(frame_y)
        frame = cv2.addWeighted(frame_x,1,frame_y,1,0)
    else:
        frame = cv2.Laplacian(frame, cv2.CV_8U, ksize=3)
    
    cv2.imshow('notebook',frame)
    out.write(frame) 
    
    if num == 10:
        num =0
    num+=1
    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

# 이미지 색상, 채도, 명도
- 색상
1. cv2.cvtColor(이미지 프레임,cv2.COLOR_RGB2HSV)
- h,s,v = cv2.split(frame) 채널을 분리할 수 있다. 
- h : 색상( 색의 질 ) 0 ~ 360 빨강 : 0 블루 : 120 초록 : 60
- s : 채도 ( 색의 선명도 ) 0 ~ 1
- v : 명도 ( 밝기 ) 0 ~ 1


```python
cap = cv2.VideoCapture(0)
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test3.avi',fourcc,fps,(width,height))
while cap.isOpened():
    
    re,frame = cap.read()    
    hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
    # define range of blue color in HSV
    lower_blue = np.array([100,50,50])
    upper_blue = np.array([130,255,255])
    
    # Treshold the HSV image to get only blue
    mask = cv2.inRange(hsv,lower_blue,upper_blue)
    
    #bitwise_AND mask and original mask
    res = cv2.bitwise_and(frame,frame,mask = mask)
    cv2.imshow('notebook',res)
    out.write(res) 
    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

# 채널 분리
1. cv2.split(이미지 프레임) ==> bgr hsv 가능

```python
cap = cv2.VideoCapture(0)
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test3.avi',fourcc,fps,(width,height))
while cap.isOpened():
    
    re,frame = cap.read()    
    rgb = cv2.cvtColor(frame,cv2.COLOR_BGR2RGB)
    r,g,b = cv2.split(rgb)
    
    h,w,c = frame.shape
    zeros = np.zeros((h,w,1),dtype = np.uint8)
    bgz = cv2.merge((b,g,zeros))
    cv2.imshow('notebook',bgz) # blue, green 채널 출력
    
    out.write(bgz) 
    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

# 이미지 도형, 글자 생성
1. line(프레임 이미지,(시작x,시작y),(끝x,끝y) ,(B,G,R),굵기,선 종류)
2. circle(프레임 이미지,(중심x,중심y),반지름,(B,G,R),색채움, 선종류)
3. rectangle(프레임 이미지,(시작x,시작y),(끝x,끝y) ,(B,G,R),굵기,선 종류)
4. polylines(프레임 이미지, [꼭지점 좌표], (B,G,R),선굵기)
5. fillPoly(프레임 이미지, [꼭지점 좌표], (B,G,R), 선종류)
6. putText(프레임 이미지,"쓸말",(위치x,위치y),폰트,폰트크기,(B,G,R),굵기)

```python
cap = cv2.VideoCapture(0)
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test3.avi',fourcc,fps,(width,height))
while cap.isOpened():
    
    re,frame = cap.read()
    cv2.line(frame, (100, 100), (200, 200), (0, 0, 255), 3, cv2.LINE_AA)
    cv2.circle(frame, (300, 300), 50, (0, 255, 0), cv2.FILLED, cv2.LINE_4)
    cv2.rectangle(frame, (0, 0), (300, 400), (255, 0, 0), 5, cv2.LINE_8)
    pts1 = np.array([[10, 50], [30, 50], [20, 60]])
    pts2 = np.array([[60, 50], [80, 50], [70, 60]])
    cv2.polylines(frame, [pts1], True, (0, 255, 255), 2)
    cv2.fillPoly(frame, [pts2], (255, 0, 255), cv2.LINE_AA)
    cv2.putText(frame, "TEST", (400, 200), cv2.FONT_HERSHEY_COMPLEX, 2, (255, 255, 255), 3)
    cv2.imshow('notebook',frame)
    out.write(frame) 
    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

# 이미지 윤곽선
1. cv2.findContour(이진이미지,검색알고리즘,근사화방법)
- contours, hierarchy = cv2.findContours(binary, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)
2. cv2.goodFeaturesToTrack(입력 이미지, 코너 최댓값, 코너 품질, 최소 거리, 마스크, 블록 크기, 해리스 코너 검출기 유/무, 해리스 코너 계수)
입력 이미지는 8비트 또는 32비트의 단일 채널 이미지를 사용합니다.    

코너 최댓값은 검출할 최대 코너의 수를 제한합니다. 코너 최댓값보다 낮은 개수만 반환합니다.   

코너 품질은 반환할 코너의 최소 품질을 설정합니다. 코너 품질은 0.0 ~ 1.0 사이의 값으로 할당할 수 있으며, 일반적으로 0.01 ~ 0.10 사이의 값을 사용합니다.   

최소 거리는 검출된 코너들의 최소 근접 거리를 나타내며, 설정된 최소 거리 이상의 값만 검출합니다.

마스크는 입력 이미지와 같은 차원을 사용하며, 마스크 요솟값이 0인 곳은 코너로 계산하지 않습니다.

블록 크기는 코너를 계산할 때, 고려하는 코너 주변 영역의 크기를 의미합니다.

해리스 코너 검출기 유/무는 해리스 코너 검출 방법 사용 여부를 설정합니다.

해리스 코너 계수는 해리스 알고리즘을 사용할 때 할당하며 해리스 대각합의 감도 계수를 의미합니다.

Tip : 코너 품질에서 가장 좋은 코너의 강도가 1000이고, 코너 품질이 0.01이라면 10 이하의 코너 강도를 갖는 코너들은 검출하지 않습니다.
Tip : 최소 거리의 값이 5일 경우, 거리가 5 이하인 코너점은 검출하지 않습니다.

```python
cap = cv2.VideoCapture(0)
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test3.avi',fourcc,fps,(width,height))
while cap.isOpened():
    
    re,frame = cap.read()
    gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
    ret, binary = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY)
    binary = cv2.bitwise_not(binary)

    contours, hierarchy = cv2.findContours(binary, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)
    # 검색 알고리즘 cv2.RETR_CCOMP // 근사화 방법 cv2.CHAIN_APPROX_NONE

    for i in range(len(contours)):
        cv2.drawContours(frame, [contours[i]], 0, (0, 0, 255), 2)
        cv2.putText(frame, str(i), tuple(contours[i][0][0]), cv2.FONT_HERSHEY_COMPLEX, 0.8, (0, 255, 0), 1)
#         print(i, hierarchy[0][i])
    
    cv2.imshow('notebook',frame)
    out.write(frame) 
    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test3.avi',fourcc,fps,(width,height))
while cap.isOpened():
    
    re,frame = cap.read()
    dst = frame.copy()

    gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)

    corners = cv2.goodFeaturesToTrack(gray, 100, 0.01, 5, blockSize=3, useHarrisDetector=True, k=0.03)
    # 코너 품질 ( 0 ~ 1, 일반적으로는 0.01 ~ 0.1  ) // 100과 0.01 만 잘 바꾸면된다.
    for i in corners:
        cv2.circle(dst, tuple(i[0]), 3, (0, 0, 255), 2)
    
    cv2.imshow('notebook',dst)
    out.write(dst) 
    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```

# 외곽선 검출
gray + threshold+ findContours+ convexHull + drawContours

1. convexHull : 윤곽선(points, contours)의 경계면을 둘러싸는 다각형을 구하는 알고리즘

- cv2.convexHull(윤곽선, 방향)
2. cv2.findContours(이진화 이미지, 검색 방법, 근사화 방법)

반환값으로 윤곽선, 계층 구조를 반환합니다.

윤곽선은 Numpy 구조의 배열로 검출된 윤곽선의 지점들이 담겨있습니다.

계층 구조는 윤곽선의 계층 구조를 의미합니다. 각 윤곽선에 해당하는 속성 정보들이 담겨있습니다.

3. cv2.drawContours(이미지, [윤곽선], 윤곽선 인덱스, (B, G, R), 두께, 선형 타입)

윤곽선은 검출된 윤곽선들이 저장된 Numpy 배열입니다.

윤곽선 인덱스는 검출된 윤곽선 배열에서 몇 번째 인덱스의 윤곽선을 그릴지를 의미합니다.

Tip : 윤곽선 인덱스를 0으로 사용할 경우 0 번째 인덱스의 윤곽선을 그리게 됩니다. 하지만, 윤곽선 인수를 대괄호로 다시 묶을 경우, 0 번째 인덱스가 최댓값인 배열로 변경됩니다.

Tip : 동일한 방식으로 [윤곽선], 0과 윤곽선, -1은 동일한 의미를 갖습니다. (-1은 윤곽선 배열 모두를 의미)

```python
cap = cv2.VideoCapture(0)
height = 480
width = 640
fourcc = cv2.VideoWriter_fourcc(*'DIVX') # 코덱설정 
fps = cap.get(cv2.CAP_PROP_FPS)
out = cv2.VideoWriter('test3.avi',fourcc,fps,(width,height))
while cap.isOpened():
    
    re,frame = cap.read()
    dst = frame.copy()

    gray = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY)
    ret, binary = cv2.threshold(gray, 120, 255, cv2.THRESH_BINARY_INV)

    contours, hierarchy = cv2.findContours(binary, cv2.RETR_CCOMP, cv2.CHAIN_APPROX_NONE)

    for i in contours:
        hull = cv2.convexHull(i, clockwise=True)
        cv2.drawContours(dst, [hull], 0, (0, 0, 255), 2)
    cv2.imshow('notebook',dst)
    out.write(dst) 
    if cv2.waitKey(1) >0 :
        break
cap.release()
out.release()
cv2.destroyAllWindows()
```
