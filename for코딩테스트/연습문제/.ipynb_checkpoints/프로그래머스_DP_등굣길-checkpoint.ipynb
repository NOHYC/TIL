```python
def solution(m, n, puddles):
    maps = [[1 if [i+1,j+1] in puddles else 0 for i in range(m)] for j in range(n)]
    new =[]
    for i,lis in enumerate(maps):
        new_in = []
        for j in range(len(lis)):
            if i == 0 and j == 0:
                new_in.append(1)
                continue
            if i == 0:
                if maps[i][j] == 1:
                    new_in.append(0)
                elif j>0:
                    if new_in[j-1] == 0:
                        new_in.append(0)
                    else:
                        new_in.append(1)
            elif j ==  0:
                if maps[i][j] == 1:
                    new_in.append(0)
                elif i > 0:
                    if new[i-1][j] == 0:
                        new_in.append(0)
                    else:
                        new_in.append(1)
            else:
                if maps[i][j] == 1:
                    new_in.append(0)
                else:
                    new_in.append(new[i-1][j] + new_in[j-1])
        new.append(new_in)
    return new[n-1][m-1]%1000000007
```
