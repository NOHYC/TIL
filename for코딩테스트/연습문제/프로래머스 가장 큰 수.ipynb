```python
def solution(numbers):
    qs = qsort(numbers)[::-1]
    answer = ''
    for i in qs:
        answer += str(i)
    return answer

def qsort(lis):
    if len(lis) <= 1:
        return lis
    left,right = [], []
    pivot = lis[0]
    str_pivot = str(pivot)
    
    for i in range(1,len(lis)):
        str_num = str(lis[i])
        len_str_num = len(str_num)
        num = 0
        while True:
            if len_str_num == len(str_pivot):
                if len_str_num == num:
                    right.append(lis[i])
                    break
                if int(str_pivot[num]) > int(str_num[num]):
                    left.append(lis[i])
                    break
                elif int(str_pivot[num]) < int(str_num[num]):
                    right.append(lis[i])
                    break
                else:
                    num+=1
            elif len_str_num > len(str_pivot):
                if len_str_num == num:
                    right.append(lis[i])
                    break
                if len(str_pivot)-1 >= num:
                    if int(str_pivot[num]) > int(str_num[num]):
                        left.append(lis[i])
                        break
                    elif int(str_pivot[num]) < int(str_num[num]):
                        right.append(lis[i])
                        break
                        
#                 print(solution([340,3403]), 3403403)
#                 print(solution([340,3402]), 3403402)
#                 print(solution([340,3405]), 3405340)
                elif len(str_pivot)-1 < num:# p3 i30
                    back = int(str_pivot[len(str_pivot)-1]) + int(str_num[0]) # 3
                    front = int(str_pivot[0]) +int(str_num[num]) # 6
                    if back - front > 0:
                        left.append(lis[i])
                        break
                    elif back - front < 0:
                        right.append(lis[i])
                        break 
#                     else:
                        
                num+=1

            else:
                if len(str_pivot) == num :
                    left.append(lis[i])
                    break
                if len_str_num - 1 >= num :
                    if int(str_pivot[num]) > int(str_num[num]):
                        left.append(lis[i])
                        break
                    elif int(str_pivot[num]) < int(str_num[num]):
                        right.append(lis[i])
                        break
#                         print(solution([3054,305]), 3054305)
#                         print(solution([3044,304]), 3044304)
                elif len_str_num - 1 < num : #p30 i3
                    back = int(str_pivot[num]) + int(str_num[0]) # 303
                    front = int(str_pivot[0]) +int(str_num[len_str_num-1]) # 330
                    if back-front > 0 :
                        left.append(lis[i])
                        break
                    elif back - front < 0:
                        right.append(lis[i])
                        break
                num+=1
    return qsort(left) + [pivot] + qsort(right)
```
