```python
class Node(object):
    def __init__(self,data = None,pointer = None):
        self.data = data
        self.pointer = None
        
    def getData(self):
        return self.data
    
    def getNext(self):
        return self.pointer
    
    def setData(self,new_data):
        self.data = new_data
        
    def setNext(self,newpointer):
        self.pointer = newpointer
        
```

```python
class linked_list(object):# 선입 선출
    def __init__(self):
        self.head = None
        self.tail = None
        self.length = 0
        
    def print_node(self):
        node = self.head
        while node:
            print(node.data)
            node = node.pointer
            
    
    def _addFirst(self,value):
        self.length = 1
        node = Node(value)
        self.head = node
        self.tail = node
        
    def _delFirst(self):
        self.length = 0
        self.head = None
        self.tail = None
        print('연결리스트가 비었습니다.')
        
    def _add(self,value):
        self.length +=1
        node = Node(value)
        if self.tail:
            self.tail.pointer = node
        self.tail = node
    
    def addNode(self,value):
        if not self.head:
            self._addFirst(value)
        else:
            self._add(value)
    
    def _findind(self,index):
        prev = None
        node = self.head
        i = 0
        while node and i < index:
            prev = node
            node = node.pointer
            i +=1
        return node , prev , i
    
    def _findval(self,value):
        prev = None
        node = self.head
        found = False
        while node and not found:
            if node.data == value:
                found = True
            else:
                prev = node
                node = node.pointer
        return node,prev,found
    
    def delNodeind(self,index):
        if not self.head or not self.head.pointer:
            self._delFirst()
        else:
            node,prev,i = self._findind(index)
            if i == index and node:
                self.length -=1
                if i ==0 or not prev:
                    self.head = node.pointer
                    self.tail = node.pointer
                else:
                    prev.pointer = node.pointer
            else:
                print(f"{index}에 해당하는 노드가 없습니다.")
    def delNodeval(self,value):
        if not self.head or not self.head.pointer:
            self._delFirst()
        else:
            node,prev,found = self._findval(value)
            if value == node.data  and node:
                self.length -= 1
                if found == 0 or not prev:
                    self.head = node.pointer
                    self.tail = node.pointer
                else:
                    prev.pointer = node.pointer
            else:
                print(f'{value}에 해당되는 노드가 없습니다.')
```
